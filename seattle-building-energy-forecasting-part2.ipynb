#### Import data

```python
with open("data/part1.pkl", "rb") as f:
    my_unpickler = pickle.Unpickler(f)
    data = my_unpickler.load()
```

```python
data.head(20)
```

#### NaN by column

```python
pd.DataFrame(data.isna().sum() / len(data)).sort_values(by=0, ascending=False)
```

#### Remove column with too many NaN

```python
col_to_remove = []
for col in data.columns.drop("ENERGYSTARScore"):
    if (data[col].isna().sum() / len(data) > 0.2):
        col_to_remove.append(col)
```

```python
data.drop(col_to_remove, axis=1, inplace=True)
```

```python
pd.DataFrame(data.isna().sum() / len(data)).sort_values(by=0, ascending=False)
```

#### Drop useless features

```python
cols_to_drop = ["OSEBuildingID", "PropertyName", "City", "State", "TaxParcelIdentificationNumber", "ListOfAllPropertyUseTypes"]
```

```python
data.drop(cols_to_drop, axis=1, inplace=True)
```

```python
pd.DataFrame(data.isna().sum() / len(data)).sort_values(by=0, ascending=False)
```

#### Check correlated features

```python
data.corr()
```

```python
temp = data.corr()
checked = []
for col in temp.columns:
    for ind in temp.index:
        if (((ind, col) not in checked) and (ind != col)):
            checked.append((col, ind))
            if temp.loc[ind, col] > 0.75:
                print(f"{ind} - {col} - {temp.loc[ind, col]}")
```

#### Remove correlated features

```python
# TotalGHGEmissions
cols_to_remove = ["SiteEUIWN(kBtu/sf)", "SourceEUI(kBtu/sf)", "SourceEUIWN(kBtu/sf)", "SourceEUI(kBtu/sf)", "SourceEUIWN(kBtu/sf)", "LargestPropertyUseTypeGFA", "PropertyGFABuilding(s)", "SiteEnergyUseWN(kBtu)", "PropertyGFATotal", "Electricity(kWh)", "SteamUse(kBtu)", "NaturalGas(therms)", "GHGEmissionsIntensity", "SiteEUI(kBtu/sf)", "NaturalGas(kBtu)", "Electricity(kBtu)"]
data.drop(cols_to_remove, axis=1, inplace=True)
```

```python
temp = data.corr()
checked = []
for col in temp.columns:
    for ind in temp.index:
        if (((ind, col) not in checked) and (ind != col)):
            checked.append((col, ind))
            if temp.loc[ind, col] > 0.7:
                print(f"{ind} - {col} - {temp.loc[ind, col]}")
```

```python
pd.DataFrame(data.isna().sum() / len(data)).sort_values(by=0, ascending=False)
```

#### Remove rows with no target

```python
data = data[~data["TotalGHGEmissions"].isna()]
```

```python
pd.DataFrame(data.isna().sum() / len(data)).sort_values(by=0, ascending=False)
```
